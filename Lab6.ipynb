```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols =['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F_Undergrad', 'P_Undergrad', 'Outstate', 'Room_Board', 'Books', 'Personal', 'PhD', 'Terminal', 'S_F_Ratio', 'perc_alumni', 'Expend', 'Grad_Rate'], outputCol='features')
from pyspark.ml.feature import StringIndexer
output = assembler.transform(treedata)
indexer = StringIndexer(inputCol="Private", outputCol="PrivateIndex")
output_fixed = indexer.fit(output).transform(output)

# Get the complete dataset selected and split ready for trainig models
final_data = output_fixed.select('features', 'PrivateIndex') # Used to select a subset of columns from a DataFrame.
trainData, testData = final_data.randomSplit([0.8, 0.2]) # randomSplit() is used to split a DataFrame into multiple subsets randomly.  The split ([0.7, 0.3]) determines the training and test data size split
final_data.select('features').show()
```

```python
#Decision tree classifier
from pyspark.sql import SparkSession
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline

# Initialize Spark session (if not already running, though it should be from previous cells)
# spark = SparkSession.builder.appName("DecisionTreeClassifiers").getOrCreate() # This line might be redundant if spark is already initialized

# Use the trainData and testData prepared in the previous cell
train_data = trainData
test_data = testData

# Define feature assembler to ensure the 'features' column is a vector
# Note: Your previous steps already created a 'features' vector,
# but this assembler can be useful in the pipeline if you were combining other columns.
# Since 'features' is already the correct format, this assembler stage might be simplified
# or even removed depending on the exact pipeline structure you intend.
# For now, I'll keep it but ensure it uses the existing 'features' column.
# If 'features' is already a VectorUDT, the assembler might not be strictly necessary
# but keeping it here demonstrates how to include it in a pipeline.
assembler = VectorAssembler(inputCols=["features"], outputCol="assembledFeatures")


# 1. Basic Decision Tree Classifier (default parameters)
# Use 'PrivateIndex' as the label column
dt_basic = DecisionTreeClassifier(labelCol="PrivateIndex", featuresCol="assembledFeatures", seed=1234)
pipeline_basic = Pipeline(stages=[assembler, dt_basic])
model_basic = pipeline_basic.fit(train_data)
predictions_basic = model_basic.transform(test_data)

# 2. Decision Tree with Entropy impurity
# Use 'PrivateIndex' as the label column
dt_entropy = DecisionTreeClassifier(
    labelCol="PrivateIndex",
    featuresCol="assembledFeatures",
    impurity="entropy",
    seed=1234
)
pipeline_entropy = Pipeline(stages=[assembler, dt_entropy])
model_entropy = pipeline_entropy.fit(train_data)
predictions_entropy = model_entropy.transform(test_data)

# 3. Tuned Decision Tree (Gini impurity, maxDepth=5)
# Use 'PrivateIndex' as the label column
dt_tuned = DecisionTreeClassifier(
    labelCol="PrivateIndex",
    featuresCol="assembledFeatures",
    impurity="gini",
    maxDepth=5,
    seed=1234
)
pipeline_tuned = Pipeline(stages=[assembler, dt_tuned])
model_tuned = pipeline_tuned.fit(train_data)
predictions_tuned = model_tuned.transform(test_data)

# Evaluate test error for each model
# Use 'PrivateIndex' as the label column
evaluator = MulticlassClassificationEvaluator(
    labelCol="PrivateIndex", predictionCol="prediction", metricName="accuracy"
)

# Calculate accuracy and test error
accuracy_basic = evaluator.evaluate(predictions_basic)
test_error_basic = 1.0 - accuracy_basic
accuracy_entropy = evaluator.evaluate(predictions_entropy)
test_error_entropy = 1.0 - accuracy_entropy
accuracy_tuned = evaluator.evaluate(predictions_tuned)
test_error_tuned = 1.0 - accuracy_tuned

# Print results
print(f"Basic Decision Tree - Test Accuracy: {accuracy_basic:.4f}, Test Error: {test_error_basic:.4f}")
print(f"Entropy Decision Tree - Test Accuracy: {accuracy_entropy:.4f}, Test Error: {test_error_entropy:.4f}")
print(f"Tuned Decision Tree (Gini, maxDepth=5) - Test Accuracy: {accuracy_tuned:.4f}, Test Error: {test_error_tuned:.4f}")

# Stop Spark session (optional, depending on if you plan to run more Spark code)
# spark.stop()
```

```python
# Get the trained Decision Tree model from the pipeline
# The last stage in the pipeline is the Decision Tree Classifier
dt_model = model_tuned.stages[-1]

# Print the description of the tree model
print(dt_model.toDebugString)
```

K-Means clustering mode from a tutorial in https://spark.apache.org/docs/latest/ml-clustering.html#k-means

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

#loads data
dataset = spark.read.format("libsvm").load("/content/drive/MyDrive/Colab Notebooks/kmeans_data.txt")

# Trains a k-means model.
kmeans = KMeans().setK(2).setSeed(1)
model = kmeans.fit(dataset)

# Make predictions
predictions = model.transform(dataset)

# Evaluate clustering by computing Silhoutte score
evaluator = ClusteringEvaluator()

silhoutte = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance =" +str(silhoutte))

# Show the result
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
  print(center)
```
