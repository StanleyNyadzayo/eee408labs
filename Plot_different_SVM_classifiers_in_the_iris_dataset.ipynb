```python
import matplotlib.pyplot as plt

from sklearn import datasets, svm
from sklearn.inspection import DecisionBoundaryDisplay

# import data for learning purposes
iris = datasets.load_iris()
# Take the first 2 features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# Create an instance of SVM and fit out data.

C = 1.0 # SVM regularisation parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)

# Add titles for each model
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)


# Set- 3x3 grid for plotting.
fig, sub = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Define X0 and X1 for plotting
X0 = X[:, 0]
X1 = X[:, 1]
#X0, X1= X[:, 0], X[:,1]

for clf in models:
    clf.fit(X, y)

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=25, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(()) # Corrected typo here
    ax.set_title(title)

plt.show()
```
