```python
import numpy as np
import pandas as pd

data = {'sentence': ['One year into lockdown done!',
                     'AI 2 Module is almost complete and Easter is near',
                     'Lets see what these tokens are all about']}

testSents = pd.DataFrame(data)
print(testSents)
```

```python
common = pd.Series(' '.join(testSents['sentence']).split()).value_counts()[:40]
common
#uncommon = pd.Series(' '.join(testSents['sentence']).split()).value_counts(ascending=True)[:40]
#uncommon
```

```python
testSents['sentence'] = testSents['sentence'].str.lower()
print(testSents)
```

```python
import re
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python
testSents['sentence'] = testSents['sentence'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words]))
display(testSents)
```

```python
from nltk.tokenize import word_tokenize
testSents['tokenized_sentence'] = testSents['sentence'].apply(word_tokenize)
display(testSents)
```
